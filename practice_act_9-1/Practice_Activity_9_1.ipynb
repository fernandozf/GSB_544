## 1. Logistic Regression

Fit a Logistic Regression using only `age` and `chol` (cholesterol) as predictors.

For a 55 year old, how high would their cholesterol need to be for the doctors to predict heart disease is present?

It would need to be above 1587.96

How high for the doctors to estimate a 90% chance that heart disease is present?

It would need to be above 367.96

```python
import pandas as pd
import numpy as np
from sklearn.pipeline import Pipeline
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.compose import make_column_selector, ColumnTransformer
from sklearn.preprocessing import StandardScaler, OneHotEncoder, PolynomialFeatures
from sklearn.model_selection import train_test_split, cross_val_score, GridSearchCV
from sklearn.metrics import roc_curve, auc, confusion_matrix, accuracy_score, recall_score, precision_score, cohen_kappa_score
from plotnine import *
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis as LDA
from sklearn.svm import SVC
```

```python
ha = pd.read_csv("https://www.dropbox.com/scl/fi/0vrpdnq5asmeulc4gd50y/ha_1.csv?rlkey=ciisalceotl77ffqhqe3kujzv&dl=1")
```

```python
X = ha[['age','chol']]
y = ha['diagnosis']

model_pipeline = Pipeline([
        ("model", LogisticRegression())
    ])

fit = model_pipeline.fit(X,y)

model = model_pipeline.named_steps['model']


LRcoefficients = model.coef_[0]
LRintercept = model.intercept_[0]

coef_df = pd.DataFrame({
    'Feature': X.columns.tolist() + ['Intercept'],
    'Coefficient': LRcoefficients.tolist() + [LRintercept]
})

print(coef_df)
```

## 2. Linear Discriminant Analysis

Fit an LDA model using only `age` and `chol` (cholesterol)  as predictors.

For a 55 year old, how high would their cholesterol need to be for the doctors to predict heart disease is present?

it would need to be above 368.18

```python
X = ha[['age','chol']]
y = ha['diagnosis']

model_pipeline = Pipeline([
        ("model", LDA())
    ])

fit = model_pipeline.fit(X,y)

model = model_pipeline.named_steps['model']

LDAcoefficients = model.coef_[0]
LDAintercept = model.intercept_[0]

coef_df = pd.DataFrame({
    'Feature': X.columns.tolist() + ['Intercept'],
    'Coefficient': LDAcoefficients.tolist() + [LDAintercept]
})

print(coef_df)
```

## 3. Support Vector Classifier

Fit an SVC model using only `age` and `chol` as predictors.  Don't forget to tune the regularization parameter.

For a 55 year old, how high would their cholesterol need to be for the doctors to predict heart disease is present?

it would need to be above 312.66

```python
X = ha[['age','chol']]
y = ha['diagnosis']

model_pipeline = Pipeline([
        ("model", SVC(kernel = 'linear'))
    ])

degree = {'model__C': np.array([.001,.01,.05,.1, .2,.5])}

gscv = GridSearchCV(model_pipeline, degree, cv = 5, scoring='accuracy')
tuning = gscv.fit(X, y)

best_model = tuning.best_estimator_

fit = best_model.fit(X,y)

model = best_model.named_steps['model']


SVCcoefficients = model.coef_[0]
SVCintercept = model.intercept_[0]

coef_df = pd.DataFrame({
    'Feature': X.columns.tolist() + ['Intercept'],
    'Coefficient': SVCcoefficients.tolist() + [SVCintercept]
})

print(coef_df)
```

## 4. Comparing Decision Boundaries

Make a scatterplot of `age` and `chol`, coloring the points by their true disease outcome.  Add a line to the plot representing the **linear separator** (aka **decision boundary**) for each of the three models above.

```python
SVC_age = SVCcoefficients[0]
SVC_chol = SVCcoefficients[1]
SVC_inter = SVCintercept

LDA_age = LDAcoefficients[0]
LDA_chol = LDAcoefficients[1]
LDA_inter = LDAintercept

LR_age = LRcoefficients[0]
LR_chol = LRcoefficients[1]
LR_inter = LRintercept

SVCslope = -SVC_age/SVC_chol
LDAslope = -LDA_age/LDA_chol
LRslope = -LR_age/LR_chol

SVCinter = -SVC_inter/SVC_chol
LDAinter = -LDA_inter/LDA_chol
LRinter = -LR_inter/LR_chol
```

0 = w1x1 +w2x2 + c  
-w2x2 = w1x1 + c  
x2 = - w1/w2 x1 - c/w2  

```python
plot = (ggplot(ha) +
        geom_abline(aes(slope=LRslope, intercept=LRinter), color='blue', linetype='dashed', size=1) +
        geom_abline(aes(slope=LDAslope, intercept=LDAinter), color='green', linetype='dashed', size=1.2) +
        geom_abline(aes(slope=SVCslope, intercept=SVCinter), color='red', linetype='dashed', size=1) +
        geom_point(aes(x='age', y='chol', color='diagnosis'), size=2) +
        labs(title='Decision boundaries', x='Age', y='Chol') +
        theme_minimal()
        )

plot += geom_text(aes(x=30, y=LDAslope * 30 + LDAinter), label="LDA", color="green", size=10, ha='left')
plot += geom_text(aes(x=35, y=LRslope * 35 + LRinter), label="LR", color="blue", size=10, ha='left')
plot += geom_text(aes(x=40, y=SVCslope * 40 + SVCinter), label="SVC", color="red", size=10, ha='left')

plot
```
