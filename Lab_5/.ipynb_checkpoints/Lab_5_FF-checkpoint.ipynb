# **Part Two: Simple Linear Models**

## 1. Construct a simple linear model to predict the insurance charges from the beneficiarys age. Discuss the model fit, and interpret the coefficient estimates.

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler, OneHotEncoder, PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score, mean_squared_error
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import FunctionTransformer
```

```python
## splitting x_train and X_test for the only time
X = health_insur.drop('charges', axis=1)
y = health_insur['charges']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25)

def return_same(X):
  return X

pass_variable = FunctionTransformer(return_same)
```

```python
# creating collum tranformer to use only variable age
ct = ColumnTransformer(
  [('untouched_exp_var', pass_variable, ['age'])
  ],
  remainder = "drop"
)

#creating pipeline
lr_pipe = Pipeline(
    [('preprocessing', ct),
        ('linear_reg', LinearRegression())]
)

#fitting it
fitted_lr_pipe = lr_pipe.fit(X_train,y_train)


y_preds = fitted_lr_pipe.predict(X_test)

r2_part11 = r2_score(y_test, y_preds)
mse_part11 = mean_squared_error(y_test, y_preds)

coeff = fitted_lr_pipe.named_steps['linear_reg'].coef_

r2_part11, mse_part11, coeff
```

if we are trying to predict the insurance costs based on age, we dont get a very good result as we can see our r squared of only .11.

The coefficient of the explanatory age tells us that for one year increase in age, the predicted insurance cost will increase by 214.61 dollars.

## 2.Make a model that also incorporates the variable sex. Report your results.

```python
ct = ColumnTransformer(
  [('untouched_exp_var', pass_variable, ['age']),
   ("dummify", OneHotEncoder(), ['sex'])
   ],
  remainder = "drop"
)


lr_pipe = Pipeline(
    [('preprocessing', ct),
        ('linear_reg', LinearRegression())]
)


fitted_lr_pipe = lr_pipe.fit(X_train,y_train)


y_preds = fitted_lr_pipe.predict(X_test)

r2_part12 = r2_score(y_test, y_preds)
mse_part12 = mean_squared_error(y_test, y_preds)

r2_part12, mse_part12
```

we can actually see that by adding the variable sex, the model actually performed slightly worse with an r squared of .107

## 3.Now make a model that does not include sex, but does include smoker. Report your results

```python
ct = ColumnTransformer(
  [('untouched_exp_var', pass_variable, ['age']),
   ("dummify", OneHotEncoder(), ['smoker'])
   ],
  remainder = "drop"
)


lr_pipe = Pipeline(
    [('preprocessing', ct),
        ('linear_reg', LinearRegression())]
)


fitted_lr_pipe = lr_pipe.fit(X_train,y_train)


y_preds = fitted_lr_pipe.predict(X_test)

r2_part13 = r2_score(y_test, y_preds)
mse_part13 = mean_squared_error(y_test, y_preds)

r2_part13, mse_part13
```

Now when we add soke, we see that our model performed a lot better. This is represented by the new r squared of .739. We can conclude that whether you are a smoker or not probably plays a huge role in determining your insurance costs

## 4.Which model (Q2 or Q3) do you think better fits the data? Justify your answer by calculating the MSE for each model, and also by comparing R-squared values.

```python
#do not display as scientific notation, instead use two decimal places
pd.set_option('display.float_format', lambda x: '%.2f' % x)

table1 = pd.DataFrame({
    'Model': ['Model Q2', 'Model Q3'],
    'R^2 Score': [r2_part12, r2_part13],
    'MSE': [mse_part12, mse_part13]
})

table1
```

By comparing the results, we can see that the model from question 3 including smoker performed way better with an r squared of .74 and mse of 41230628 compared to model with sex and age with r squared of .11 and mse of 140898236. We can conclude that smoker plays a much bigger role in determining insurance costs than sex.

# **Part Four: New data**

For each model, fit the model on the original data.

Then, use the fitted model to predict on the new data.

Report the MSE for each models new predictions. Based on this, which is the best model to use?

Make a plot showing the residuals of your final chosen model.

```python
new_insur = pd.read_csv("https://www.dropbox.com/s/sky86agc4s8c6qe/insurance_costs_2.csv?dl=1")
Xnew = new_insur.drop('charges', axis=1)
ynew = new_insur['charges']
```

## Model 1 - only age as predictor

```python
# creating collum tranformer to use only variable age
ct = ColumnTransformer(
  [('untouched_exp_var', pass_variable, ['age'])
  ],
  remainder = "drop"
)

#creating pipeline
lr_pipe = Pipeline(
    [('preprocessing', ct),
        ('linear_reg', LinearRegression())]
)


fitted_lr_pipe = lr_pipe.fit(X,y)


y_preds = fitted_lr_pipe.predict(Xnew)

r2_part31 = r2_score(ynew, y_preds)
mse_part31 = mean_squared_error(ynew, y_preds)
```

## Model 2 - age and bmi as predictors

```python
ct = ColumnTransformer(
  [('untouched_exp_var', pass_variable, ['age','bmi'])
  ],
  remainder = "drop"
)

#creating pipeline
lr_pipe = Pipeline(
    [('preprocessing', ct),
        ('linear_reg', LinearRegression())]
)


fitted_lr_pipe = lr_pipe.fit(X,y)


y_preds = fitted_lr_pipe.predict(Xnew)

r2_part32 = r2_score(ynew, y_preds)
mse_part32 = mean_squared_error(ynew, y_preds)
```

## Model 3 - age, bmi and smoker as predictors

```python
ct = ColumnTransformer(
  [('untouched_exp_var', pass_variable, ['age','bmi']),
   ("dummify", OneHotEncoder(), ['smoker'])
  ],
  remainder = "drop"
)

#creating pipeline
lr_pipe = Pipeline(
    [('preprocessing', ct),
        ('linear_reg', LinearRegression())]
)


fitted_lr_pipe = lr_pipe.fit(X,y)


y_preds = fitted_lr_pipe.predict(Xnew)

r2_part33 = r2_score(ynew, y_preds)
mse_part33 = mean_squared_error(ynew, y_preds)
```

## Model 4 - age and bmi with both interaction with smoker



```python

ct_dummies = ColumnTransformer(
  [("dummify", OneHotEncoder(sparse_output = False), ["smoker"])],
  remainder = "passthrough"
).set_output(transform = "pandas")


ct = ColumnTransformer(
  [('untouched_exp_var', pass_variable, ['remainder__age','remainder__bmi']),
   ("interaction", PolynomialFeatures(interaction_only = True), ["remainder__age", "dummify__smoker_yes"]),
   ("interaction2", PolynomialFeatures(interaction_only = True), ["remainder__bmi", "dummify__smoker_yes"])
  ],
  remainder = "drop"
)

#creating pipeline
lr_pipe = Pipeline(
    [('preprocessing1', ct_dummies),
     ('preprocessing2', ct),
        ('linear_reg', LinearRegression())]
)


fitted_lr_pipe = lr_pipe.fit(X,y)


y_preds4 = fitted_lr_pipe.predict(Xnew)

r2_part34 = r2_score(ynew, y_preds4)
mse_part34 = mean_squared_error(ynew, y_preds4)
```

## Model 5 - age, bmi, smoker and both quantitative variables interacting with smoker

```python
ct_dummies = ColumnTransformer(
  [("dummify", OneHotEncoder(sparse_output = False), ["smoker"])],
  remainder = "passthrough"
).set_output(transform = "pandas")


ct = ColumnTransformer(
  [('untouched_exp_var', pass_variable, ['remainder__age','remainder__bmi','dummify__smoker_yes']),
   ("dummify2", OneHotEncoder(), ['remainder__region']),
   ("interaction", PolynomialFeatures(interaction_only = True), ["remainder__age", "dummify__smoker_yes"]),
   ("interaction2", PolynomialFeatures(interaction_only = True), ["remainder__bmi", "dummify__smoker_yes"])
  ],
  remainder = "drop"
)

#creating pipeline
lr_pipe = Pipeline(
    [('preprocessing1', ct_dummies),
     ('preprocessing2', ct),
        ('linear_reg', LinearRegression())]
)


fitted_lr_pipe = lr_pipe.fit(X,y)


y_preds = fitted_lr_pipe.predict(Xnew)

r2_part35 = r2_score(ynew, y_preds)
mse_part35 = mean_squared_error(ynew, y_preds)
```

## MSE and R2

```python
table3 = pd.DataFrame({
    'Model': ['Model Q1', 'Model Q2','Model Q3', 'Model Q4', 'Model Q5'],
    'R^2 Score': [r2_part31, r2_part32, r2_part33, r2_part34, r2_part35],
    'MSE': [mse_part31, mse_part32, mse_part33, mse_part34, mse_part35]
})

table3
```

Based on mse, the best model to use is moddel from question 4, with age and bmi both interacting with whether the person smokes or not.

## Plot of best model - model 4

```python
residuals = ynew - y_preds4

residuals_df = pd.DataFrame({
    'Predicted': y_preds4,
    'Residuals': residuals
})


residual_plot = (
    ggplot(residuals_df, aes(x='Predicted', y='Residuals')) +
    geom_point() +
    labs(x='Predicted Values', y='Residuals', title='Residuals Plot')
)

residual_plot
```

# **Part Five: Full Exploration**

Using any variables in this dataset, and any polynomial of those variables, find the model that best predicts on the new data after being fit on the original data.

Make a plot showing the residuals of your final chosen model.

```python
ct_dummies = ColumnTransformer(
  [("dummify", OneHotEncoder(sparse_output = False), ["smoker"])],
  remainder = "passthrough"
).set_output(transform = "pandas")


ct = ColumnTransformer(
  [('untouched_exp_var', pass_variable, ['remainder__bmi','dummify__smoker_yes']),
   ("interaction2", PolynomialFeatures(interaction_only = True), ["remainder__bmi", "dummify__smoker_yes"]),
   ('polynmial3', PolynomialFeatures((1,3)), ['remainder__age'])
  ],
  remainder = "drop"
)

#creating pipeline
lr_pipe = Pipeline(
    [('preprocessing1', ct_dummies),
     ('preprocessing2', ct),
        ('linear_reg', LinearRegression())]
)


fitted_lr_pipe = lr_pipe.fit(X,y)


y_preds_final = fitted_lr_pipe.predict(Xnew)

r2 = r2_score(ynew, y_preds_final)
mse = mean_squared_error(ynew, y_preds_final)

r2, mse
```

```python
residuals_final = ynew - y_preds_final

residuals_final_df = pd.DataFrame({
    'Predicted': y_preds_final,
    'Residuals': residuals_final
})


residual_plot = (
    ggplot(residuals_final_df, aes(x='Predicted', y='Residuals')) +
    geom_point() +
    labs(x='Predicted Values', y='Residuals', title='Residuals Plot')
)

residual_plot
```
