```python
def rest_locations(url):
    response = requests.get(url)
    soup = BeautifulSoup(response.content, "html.parser")
    locations = soup.find_all("li", attrs={"class": ""})

    rest_name = soup.find("meta", attrs={"property":"og:title", "name":'og:title'})
    restaurant_name = rest_name.get("content")

    rows = []

    for location in locations:
        a_text = location.get_text()
        before = re.findall(r".+(?= " + re.escape(restaurant_name) + " locations )", a_text)
        after = re.findall(r"(?<=locations ).+", a_text)

        if before and after:
            state = before[0]
            numloc = after[0]
            numloc = numloc.replace("(","").replace(")","")

            rows.append({
                "state": state,
                "Number_of_Locations": numloc
            })

    return pd.DataFrame(rows)
```

## 4. Scrape the state names and populations from this wikipedia page. Convert the state names to abbreviations and merge these data with your coffee dataset.

```python
response = requests.get("https://simple.wikipedia.org/wiki/List_of_U.S._states_by_population")
soup = BeautifulSoup(response.content, "html.parser")
result = soup.find_all("table")
results = result[0]

rows = []

for result in results.find_all("tr")[1:]:
  cells = result.find_all("td")

  state_tag = cells[2].find("a") or cells[2]
  state = state_tag.get_text()

  pop_tag = cells[3]
  pop = pop_tag.get_text(strip = True)

  rows.append({
          "State": state,
          "Population": pop
          })

population_by_state = pd.DataFrame(rows)
population_by_state['State'] = population_by_state['State'].map(stateabb)
population_by_state
```

```python
def stock_data(stock_ticker):
    URL = "https://finance.yahoo.com/quote/" + stock_ticker + "/"
    response = requests.get(URL)
    soup = BeautifulSoup(response.content, "html.parser")
    stk_price = soup.find_all("div", attrs={"id": "Lead-5-QuoteHeader-Proxy"})

    rows = []

    for price in stk_price:
        price_tag = price.find('fin-streamer', attrs={'class': "Fw(b) Fz(36px) Mb(-4px) D(ib)"})
        prices = price_tag.get_text(strip=True)

        company_tag = price.find("h1", attrs={'class': "D(ib) Fz(18px)"})
        company = company_tag.get_text(strip=True)

        rows.append({
            "Company": company,
            "Stock Price": prices
        })

    stkprices = pd.DataFrame(rows)
    return stkprices

# Function to get data for multiple stock tickers
def multiple_stock_data(stock_tickers):
    result_dfs = []

    for ticker in stock_tickers:
        result_df = stock_data(ticker)
        result_dfs.append(result_df)

    combined_df = pd.concat(result_dfs, ignore_index=True)
    return combined_df
```

```python
# analyzing percentage of each coffee chain by region
region_chain_distribution = pd.crosstab(rest_loc['Region'], rest_loc['Coffee Chain'], normalize='index') * 100

# analyzing percentage distribution by region for each chain
chain_region_distribution = pd.crosstab(rest_loc['Coffee Chain'], rest_loc['Region'], normalize='index') * 100
```

```python
from plotnine import *

plot = (ggplot(rest_loc, aes(x='Coffee Chain', y= 'Location Count', fill = 'Region'))
        + geom_bar(stat='identity', position= 'stack')
        + theme(figure_size=(20, 8))
)
plot
```

```python
chain_region_melted = chain_region_distribution.reset_index().melt(id_vars='Coffee Chain',
                                                                  value_name='Percentage',
                                                                  var_name='Region')


# Create the plot for distribution of regions for each coffee chain
plot2 = (ggplot(chain_region_melted, aes(x='Coffee Chain', y='Percentage', fill='Region'))
                           + geom_bar(stat='identity', position='stack')
                           + labs(title="Percentage Distribution of Regions by Coffee Chain",
                                  y="Percentage", x="Coffee Chain")
                           + theme(figure_size=(15, 10)))

plot2
```

### Does the distribution of each chain’s stores match population distribution, by both state/region?

As we can see in the graph below, most restaurants do not necessarily match their number of stores to toal population for that region. Starubucks does it for 3 regions but we see a lot more locations per capita on the west. There could be numerous factors affecting this decision by the chains, such as different life style or household income per region.

We can also see that the bigger chains logically have more stores per capita for all of the regions.

```python
store_per_capita_region = rest_loc.groupby(['Coffee Chain', 'Region'])['Location Count', 'Population'].sum().reset_index()

store_per_capita_region['Store per Capita'] = store_per_capita_region['Location Count'] / store_per_capita_region['Population']


plot3 = (ggplot(store_per_capita_region, aes(x='Coffee Chain', y='Store per Capita', fill='Region'))
                      + geom_bar(stat='identity', position='dodge')
                      + coord_flip()
                      + labs(title="Stores per Capita by Region",
                             y="Store Per Capita", x="Coffee Chain"))

plot3
```

```python
def rest_locations(url):
    response = requests.get(url)
    soup = BeautifulSoup(response.content, "html.parser")
    locations = soup.find_all("li", attrs={"class": ""})

    rest_name = soup.find("meta", attrs={"property":"og:title", "name":'og:title'})
    restaurant_name = rest_name.get("content")

    rows = []

    for location in locations:
        a_text = location.get_text()
        before = re.findall(r".+(?= " + re.escape(restaurant_name) + " locations )", a_text)
        after = re.findall(r"(?<=locations ).+", a_text)

        if before and after:
            state = before[0]
            numloc = after[0]
            numloc = numloc.replace("(","").replace(")","")

            rows.append({
                "state": state,
                "Number_of_Locations": numloc,
                "Restaurant_Name": restaurant_name
            })

    return pd.DataFrame(rows)
```
