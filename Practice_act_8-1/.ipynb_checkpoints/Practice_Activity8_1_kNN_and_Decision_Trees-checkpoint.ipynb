# Palmer Penguins Modeling

Import the Palmer Penguins dataset and print out the first few rows.

Suppose we want to predict `bill_depth_mm` using the other variables in the dataset.

**Dummify** all variables that require this.

```python
#!pip install palmerpenguins
from palmerpenguins import load_penguins
import pandas as pd
import numpy as np
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler, OneHotEncoder, PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score, mean_squared_error
from sklearn.compose import ColumnTransformer, make_column_selector
from sklearn.preprocessing import FunctionTransformer
from sklearn.neighbors import KNeighborsRegressor
from sklearn.tree import DecisionTreeRegressor
```

```python
penguins = load_penguins()
penguins.head()
```

Let's use the other variables to predict `bill_depth_mm`. Prepare your data and fit the following models on the entire dataset:

* Your best multiple linear regression model from before
* Two kNN models (for different values of K)
* A decision tree model

Create a plot like the right plot of Fig 1. in our `Model Validation` chapter with the training and test error plotted for each of your four models.

Which of your models was best?

```python
penguins = penguins.dropna()
X = penguins.drop(['bill_depth_mm'], axis =1)
y = penguins['bill_depth_mm']

X_train, X_test, y_train, y_test = train_test_split(X,y)

lr = LinearRegression()
enc = OneHotEncoder()


ct = ColumnTransformer(
    [('dummify', enc, ['species','sex','island']),
     ('standardize', StandardScaler(), ['bill_length_mm', 'flipper_length_mm'])
    ],
    remainder='drop')


pipeline_lr = Pipeline([
    ('preprocessing', ct),
     ('lr', lr)
     ])

fitted_pipeline_lr = pipeline_lr.fit(X_train,y_train)

y_test_hat = fitted_pipeline_lr.predict(X_test)
y_train_hat = fitted_pipeline_lr.predict(X_train)

r2_train_lr =  r2_score(y_train, y_train_hat)
r2_test_lr = r2_score(y_test, y_test_hat)
```

kNN

```python
ct = ColumnTransformer(
    [('dummify', enc, ['species','sex','island']),
     ('standardize', StandardScaler(), ['bill_length_mm', 'flipper_length_mm', 'body_mass_g', 'year'])
    ],
    remainder='passthrough')

pipeline_knn = Pipeline([
    ('preprocessing', ct),
     ('knn', KNeighborsRegressor(2))
     ])

fitted_pipeline_knn = pipeline_knn.fit(X_train,y_train)

y_test_hat = fitted_pipeline_knn.predict(X_test)
y_train_hat = fitted_pipeline_knn.predict(X_train)

r2_train_knn2 =  r2_score(y_train, y_train_hat)
r2_test_knn2 = r2_score(y_test, y_test_hat)
```

decision tree

```python
ct = ColumnTransformer(
    [('dummify', enc, ['species','sex','island']),
     ('standardize', StandardScaler(), make_column_selector(dtype_include=np.number))
    ],
    remainder='passthrough')


pipeline_dt = Pipeline([
    ('preprocessing', ct),
     ('dt', DecisionTreeRegressor(min_samples_leaf = 10, min_impurity_decrease = .1))
     ])

fitted_pipeline_dt = pipeline_dt.fit(X_train,y_train)

y_test_hat = fitted_pipeline_dt.predict(X_test)
y_train_hat = fitted_pipeline_dt.predict(X_train)

r2_train_dt =  r2_score(y_train, y_train_hat)
r2_test_dt = r2_score(y_test, y_test_hat)
```

```python
data_knn = {
    'Number of Neighboors': [2,3,5,10, 15, 25, 50],
    'R^2 (Train Data)': [r2_train_knn2, r2_train_knn3, r2_train_knn5, r2_train_knn10, r2_train_knn15, r2_train_knn25, r2_train_knn50],
    'R^2 (Test Data)': [r2_test_knn2, r2_test_knn3, r2_test_knn5, r2_test_knn10, r2_test_knn15, r2_test_knn25, r2_test_knn50]
}

knn = pd.DataFrame(data_knn)
knn
```

```python
def dtr2(number, test):
  ct = ColumnTransformer(
    [('dummify', enc, ['species','sex','island']),
     ('standardize', StandardScaler(), make_column_selector(dtype_include=np.number))
    ],
    remainder='passthrough')

  pipeline_dt = Pipeline([
    ('preprocessing', ct),
     ('dt', DecisionTreeRegressor(min_samples_leaf = 10, min_impurity_decrease = number))
     ])

  fitted_pipeline_dt = pipeline_dt.fit(X_train,y_train)

  y_test_hat = fitted_pipeline_dt.predict(X_test)
  y_train_hat = fitted_pipeline_dt.predict(X_train)

  r2_train_dt =  r2_score(y_train, y_train_hat)
  r2_test_dt = r2_score(y_test, y_test_hat)

  if test == "Y":
    return r2_test_dt
  else:
    return r2_train_dt
```

```python
r2_test_dt01 = dtr2(.01, test = "Y")
r2_train_dt01 = dtr2(.01, test = "N")
r2_test_dt1 = dtr2(.1, test = "Y")
r2_train_dt1 = dtr2(.1, test = "N")
r2_test_dt2 = dtr2(.2, test = "Y")
r2_train_dt2 = dtr2(.2, test = "N")
r2_test_dt5 = dtr2(.5, test = "Y")
r2_train_dt5 = dtr2(.5, test = "N")
r2_test_dt75 = dtr2(.75, test = "Y")
r2_train_dt75 = dtr2(.75, test = "N")
```

```python
data_dt = {
    'Impurity Decrease': [.01,.1,.2,.5, .75],
    'R^2 (Train Data)': [r2_train_dt01, r2_train_dt1, r2_train_dt2, r2_train_dt5, r2_train_dt75],
    'R^2 (Test Data)': [r2_test_dt01, r2_test_dt1, r2_test_dt2, r2_test_dt5, r2_test_dt75]
}

dt = pd.DataFrame(data_dt)
dt
```

```python
plot1 = (ggplot(dt,aes(x='Impurity Decrease'))
+geom_smooth(aes(y='R^2 (Test Data)', color = "'Test'"), se = False)
+geom_smooth(aes(y='R^2 (Train Data)', color = "'Train'"), se = False)
+ labs(title="R_Squared vs impurity decrease for decision tree", x="Impurity decrease", y="R^2", color="Data Type")
)
plot1, plot
```
